```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
```

```python
# generate fake data
actual = np.random.normal(size=1000)
predict = np.random.normal(size=1000)

# Calculate the point density
xy = np.vstack([actual,predict])
z = gaussian_kde(xy)(xy)
```

```python
font = {'Family': 'Times New Roman',#'Arial',
        'color': 'black',
        'weight': 'normal',
        'size':18}

fig, ax = plt.subplots(figsize=(6,6))

ax.set_title('Prediction Result', fontdict=font)

ax.scatter(actual, predict, alpha=0.7, c = z, cmap='Spectral', s=50)

ax.set_xlabel('test', fontdict=font, labelpad=None)
ax.set_ylabel('predict', fontdict=font, labelpad=None)
ax.set_xticks(np.arange(-4, 5))
ax.set_yticks(np.arange(-4, 5))
ax.set_xticklabels(np.arange(-4, 5), fontsize=14)
ax.set_yticklabels(np.arange(-4, 5), fontsize=14)
ax.set_xlim(-4,4)
ax.set_ylim(-4,4)

# Reference Line
x = range(-4,5)
y = x
ax.plot(x, y, c='k', linestyle='--', marker=None, linewidth=1)
```
